```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

def graph(x, y, color='r', fit_line=True, connect=False):
    plt.figure(figsize=(12,7))
    if connect:
        plt.plot(x, y, color + '-o')
    else:
        plt.plot(x, y, color + 'o')

    A = np.vstack([x, np.ones(len(x))]).T
    line, resid = np.linalg.lstsq(A, y, rcond=None)[:2] # fit line
    k, n = line[0], line[1]
    yfit = [k * xi + n for xi in x]
    if fit_line: plt.plot(x, yfit)
    return (line[0], line[1], resid / (len(x) * np.var(y)))
```

```python
meritve = [
    [(100, 1.7), (200, 3.8), (300, 5.8), (400, 7.8), (500, 9.7), (600, 11.2), (700, 13.2), (800, 14.3), (900, 16.2),
     (1000, 17.3), (2000, 29.6), (3000, 38.9), (4000, 46.1), (5000, 51.7), (6000, 56.6), (7000, 60.1), (8000, 62.9),
     (9000, 66.9), (10000, 68.2)],
    [(100, 1.7), (200, 3.6), (300, 5.7), (400, 7.6), (500, 9.5), (600, 11.4), (700, 12.7), (800, 14.3), (900, 16.1),
     (1000, 18.2), (2000, 30.4), (3000, 39.0), (4000, 46.2), (5000, 52.2), (6000, 56.3), (7000, 59.5), (8000, 63.3),
     (9000, 65.4)]
]
l = 1

# convert from cm to m and from ohm to kohm
for i in range(len(meritve)):
    for j in range(len(meritve[i])):
        meritve[i][j] = (meritve[i][j][0]/1000, meritve[i][j][1]/100)
        
```

```python
x = [i[0] for i in meritve[0]]
y = [i[1]/(l - i[1]) for i in meritve[0]]
k, n, err = graph(x, y)
print(k, n, err[0])
print(str(k**(-1)))
```

```python
x = [i[0] for i in meritve[1]]
y = [i[1]/(l - i[1]) for i in meritve[1]]
k, n, err = graph(x, y)
print(k, n, err[0])
print(str(k**(-1)))
```

```python
joined = sorted(meritve[0] + meritve[1])
x = [i[0] for i in joined]
y = [i[1]/(l - i[1]) for i in joined]
k, n, err = graph(x, y)
print(k, n, err[0])
print(str(k**(-1)))
```
